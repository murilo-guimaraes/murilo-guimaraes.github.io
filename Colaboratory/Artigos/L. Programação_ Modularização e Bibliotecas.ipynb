##**Introdução Teórica**

A modularização é a técnica de dividir um sistema complexo em partes menores e independentes, chamadas de módulos. Segundo Scheffer (2026), um módulo em Python é essencialmente um arquivo com extensão `.py` que agrupa definições e instruções. Essa organização não apenas facilita a manutenção do código, mas também permite a reutilização de funcionalidades através de importações, evitando a redundância e otimizando o carregamento de recursos na memória.

###**1. Importação de Módulos e Uso de Alias**

Como destacado no artigo, a sintaxe de importação permite carregar apenas o necessário. O uso de apelidos (`import modulo as m`) é uma prática comum para tornar o código mais conciso e evitar conflitos de nomes.

```python
# Importação com apelido (Alias) conforme o exemplo do artigo
import math as m

valor = 25
raiz = m.sqrt(valor)

print(f"A raiz quadrada de {valor} é {raiz}")
print(f"Valor de PI aproximado: {m.pi:.4f}")
```

####**O que observar:**

- <a>O uso do prefixo `m.` indica que estamos acessando uma funcionalidade de dentro do módulo `math`. Isso mantém o **namespace** organizado, garantindo que o interpretador saiba exatamente de onde a função `sqrt` está sendo chamada.</a>

---
###**2. O Ecossistema de Bibliotecas (Built-in vs Terceiros)**

Python é conhecido por ter "baterias inclusas". Isso significa que módulos como `random`, `os` e `datetime` já vêm instalados. Para necessidades mais específicas, utilizamos o **PyPI**, que oferece bibliotecas como o **Pandas** para manipulação de dados estruturados.

```python
import random

# Simulando a captura de dados aleatórios conforme sugerido para testes
sensores = [random.uniform(20.0, 30.0) for _ in range(5)]

print("Leituras simuladas do sensor de temperatura:")
for i, leitura in enumerate(sensores, 1):
    print(f"Sensor {i}: {leitura:.2f}°C")
```

####**O que observar:**

- <a>A biblioteca `random` é uma biblioteca **built-in** fundamental para simulações acadêmicas. Ao utilizar módulos prontos, o desenvolvedor foca na **lógica de negócio**, deixando a complexidade técnica do algoritmo de geração numérica para o módulo especializado.</a>
