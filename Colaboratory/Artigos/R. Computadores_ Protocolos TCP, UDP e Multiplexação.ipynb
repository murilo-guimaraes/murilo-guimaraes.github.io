##**Introdução Teórica**

A Camada de Transporte é responsável pela comunicação lógica entre processos de aplicação em diferentes hosts. Segundo Nunes (2017), enquanto a Camada de Rede entrega pacotes entre computadores, a de Transporte garante que os dados cheguem ao processo correto. Os dois protagonistas desta camada são o **TCP**, orientado à conexão e focado na integridade, e o **UDP**, que prioriza a velocidade sem garantias de entrega. A distinção entre esses serviços ocorre através da **Multiplexação**, utilizando números de portas lógicas para segmentar o tráfego de dados de forma eficiente.

###**1. TCP: Garantia de Conexão (Socket Stream)**

O TCP exige um "aperto de mão" (Three-way Handshake) antes de enviar dados. Vamos simular a abertura de um socket TCP para validar a prontidão da camada de transporte.

```python
import socket

def realizar_handshake(host, porta):
    try:
        # AF_INET = IPv4 | SOCK_STREAM = TCP (Orientado à conexão)
        s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        s.settimeout(3)
        s.connect((host, porta))
        print(f"[CAMADA 4] Handshake realizado com sucesso em {host}:{porta}")
        s.close()
    except Exception as e:
        print(f"[ERRO] Falha na conexão: {e}")

realizar_handshake("google.com", 80)
```

####**O que observar:**
<a>O sucesso da conexão indica que o TCP completou o ciclo de sincronização (SYN/ACK). Como discutido no artigo de Guimarães, este protocolo é indispensável para serviços onde a perda de dados é crítica, como na navegação Web (HTTP) e transferência de arquivos.</a>

###**2. Identificação de Portas e Processos**

A multiplexação permite que um único endereço IP suporte múltiplos serviços. Vamos listar as portas lógicas ativas que o sistema utiliza para direcionar os pacotes.

```python
# Verificando as conexões de transporte e suas portas (Local Address)
!netstat -tuln | head -n 10
```

####**O que observar:**
<a>As portas (ex: 80, 443, 8080) funcionam como "apartamentos" em um prédio (o IP). Como ensina Nunes, a camada de transporte usa esses números para saber exatamente qual aplicativo deve receber a informação que chegou da rede.</a>

###**3. Análise de Eficiência: TCP vs UDP**

Com base no artigo técnico, consolidamos a escolha do protocolo conforme o requisito da aplicação (Confiabilidade vs Velocidade).

```python
import pandas as pd

dados = {
    "Característica": ["Conexão", "Confiabilidade", "Velocidade", "Aplicações"],
    "TCP (Stream)": ["Orientado (Handshake)", "Alta (Recuperação)", "Moderada", "Web, E-mail"],
    "UDP (Datagram)": ["Não orientado", "Melhor esforço", "Altíssima", "VoIP, Games"]
}

df = pd.DataFrame(dados)
print(df.to_string(index=False))
```

####**O que observar:**
<a>A tabela resume o "trade-off" da Camada 4. O UDP é escolhido quando o atraso (delay) é mais prejudicial que a perda de um pacote individual, enquanto o TCP é obrigatório quando a ordem e a integridade são prioridades.</a>
