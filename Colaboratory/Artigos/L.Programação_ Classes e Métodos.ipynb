##**Introdução Teórica**

A Programação Orientada a Objetos (POO) consolidou-se como uma resposta à complexidade dos sistemas procedurais. Segundo Weisfeld (2013), enquanto as linguagens evoluem, os conceitos fundamentais da POO permanecem estáveis. A ideia central é a **Abstração**, onde utilizamos a "Classe" como um molde ou planta baixa, definindo atributos (estado) e métodos (comportamento) que serão instanciados em "Objetos" reais na memória do computador.

---
##**Objetivos**

 - **Compreender** a transição do paradigma procedural para o orientado a objetos através da analogia de "planta" versus "construção".
 - **Implementar** classes em Python utilizando o método construtor `__init__` para inicialização de atributos.
 - **Validar** comportamentos dinâmicos através de métodos, simulando cenários práticos como sistemas de HP em jogos.
 - **Diferenciar** instâncias em memória, observando como múltiplos objetos podem coexistir mantendo estados independentes.

---
###**1. Definição de Classe e Atributos**

Conforme o artigo, a classe define o que o objeto será. O método `__init__` é o responsável por "nascer" o objeto, alocando as variáveis locais (atributos) que armazenarão o estado daquela instância.

```python
class Personagem:
    def __init__(self, nome, vida):
        self.nome = nome
        self.vida = vida

    def tomar_dano(self, pontos):
        self.vida -= pontos
        print(f"{self.nome} agora tem {self.vida} HP.")

# Instanciação (A planta virando casa)
heroi = Personagem("Arqueiro", 100)
vilao = Personagem("Orc", 50)

heroi.tomar_dano(15)
```

####**O que observar:**

- <a>O uso do `self` é obrigatório para referenciar que o atributo pertence àquela instância específica. No exemplo, o dano subtraído do "Arqueiro" não afeta a vida do "Orc", demonstrando o isolamento de estado da POO.</a>

---
###**2. Métodos e Lógica de Negócio**

Métodos são funções internas da classe que definem como o objeto interage com o mundo. No artigo, destaca-se que a POO diferencia o "desenvolvimento de scripts simples" da "engenharia de software profissional" pela forma como esses comportamentos são encapsulados.

```python
class PersonagemMedio(Personagem): # Aplicando herança conforme conceitos do artigo
    def curar(self, pontos):
        self.vida += pontos
        if self.vida > 100: self.vida = 100
        print(f"{self.nome} foi curado! Vida atual: {self.vida} HP.")

mago = PersonagemMedio("Mago", 80)
mago.curar(10)
```

####**O que observar:**

- <a>A implementação de lógica dentro do método (como o limite máximo de 100 HP) demonstra o **Encapsulamento**. O mundo exterior apenas chama o método `.curar()`, mas a regra de negócio está protegida dentro da classe.</a>

---
##**Conclusão**

O experimento prático, alinhado à teoria de Weisfeld, demonstrou que a POO simplifica a manutenção e a escala do código. Ao mapear "Atributos" como variáveis de estado e "Métodos" como executores de comportamento, conseguimos modularizar sistemas complexos (como jogos e IoT). A utilização correta desses pilares é o que permite a construção de sistemas robustos e profissionalmente estruturados.
