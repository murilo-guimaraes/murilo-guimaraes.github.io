##**Introdução Teórica**

A ordenação de dados é um dos pilares da ciência da computação. O objetivo é transformar uma sequência desordenada em uma lista organizada, facilitando buscas e processamentos posteriores. Para o desenvolvedor de ADS, o estudo desses algoritmos vai além da simples organização: trata-se de entender a **Complexidade de Algoritmos**. Métodos simples como o Bubble Sort possuem um custo computacional elevado $O(N^2)$, enquanto métodos avançados como Quick e Merge Sort utilizam a estratégia de **"Dividir para Conquistar"**, atingindo $O(N \log N)$. Este relatório utiliza Python para medir o impacto dessas diferenças em cenários de execução reais.

###**1. Selection Sort (Ordenação por Seleção)**

O **Selection Sort** percorre a lista repetidamente, selecionando o menor valor restante e trocando-o com a posição atual do índice.

```python
def executar_selection_sort(lista):
    n = len(lista)
    for i in range(0, n):
        index_menor = i
        for j in range(i + 1, n):
            if lista[j] < lista[index_menor]:
                index_menor = j
        lista[i], lista[index_menor] = lista[index_menor], lista[i]
    return lista

lista_sel = [10, 9, 5, 8, 11, -1, 3]
print(f"Resultado Selection Sort: {executar_selection_sort(lista_sel)}")
```

####**O que observar:**

- <a>O **Selection Sort** sempre verifica todos os elementos à frente para encontrar o mínimo. Por isso, ele sempre operará no **pior caso**, independentemente de a lista estar organizada ou não.</a>

###**2. Bubble Sort e Insertion Sort**

O **Bubble Sort** faz os maiores valores "flutuarem" para o final, enquanto o **Insertion Sort** insere cada elemento na posição correta em relação aos valores já ordenados à esquerda.

```python
def executar_bubble_sort(lista):
    n = len(lista)
    for i in range(n - 1):
        for j in range(n - 1):
            if lista[j] > lista[j + 1]:
                lista[j], lista[j + 1] = lista[j + 1], lista[j]
    return lista

def executar_insertion_sort(lista):
    n = len(lista)
    for i in range(1, n):
        valor = lista[i]
        j = i - 1
        while j >= 0 and lista[j] > valor:
            lista[j + 1] = lista[j]
            j -= 1
        lista[j + 1] = valor
    return lista

lista_bubble = [10, 9, 5, 8, 11, -1, 3]
lista_ins = [10, 9, 5, 8, 11, -1, 3]
print(f"Bubble Sort: {executar_bubble_sort(lista_bubble)}")
print(f"Insertion Sort: {executar_insertion_sort(lista_ins)}")
```

####**O que observar:**

- <a>Diferente do Selection, o **Insertion Sort** possui excelente performance se a lista já estiver quase ordenada. Se o valor a ser inserido for grande, ele encontra sua posição rapidamente sem percorrer todo o resto.</a>

###**3. Quick Sort (Dividir para Conquistar)**

O **Quick Sort** utiliza a **recursão** e um elemento **pivô**. Ele quebra o problema em sublistas (menores e maiores que o pivô) até que a ordenação seja concluída.

```python
def executar_quicksort(lista):
    if len(lista) <= 1:
        return lista
    pivo = lista[-1]
    menores = [x for x in lista[:-1] if x <= pivo]
    maiores = [x for x in lista[:-1] if x > pivo]
    return executar_quicksort(menores) + [pivo] + executar_quicksort(maiores)

lista_quick = [10, 9, 5, 8, 11, -1, 3]
print(f"Quick Sort: {executar_quicksort(lista_quick)}")
```

####**O que observar:**

- <a>O **Quick Sort** é muito mais eficiente para grandes volumes de dados ($O(N \log N)$). Ele demonstra como a **recursividade** permite resolver problemas complexos de forma mais rápida que laços de repetição simples.</a>

###**5. Merge Sort: Estabilidade e Divisão**

O Merge Sort divide a lista em metades até que cada sublista tenha um único elemento, para então mesclá-las de forma ordenada. É um algoritmo estável e previsível.

```python
import time

def merge_sort(lista):
    if len(lista) > 1:
        meio = len(lista) // 2
        esquerda = lista[:meio]
        direita = lista[meio:]

        merge_sort(esquerda)
        merge_sort(direita)

        i = j = k = 0
        while i < len(esquerda) and j < len(direita):
            if esquerda[i] < direita[j]:
                lista[k] = esquerda[i]
                i += 1
            else:
                lista[k] = direita[j]
                j += 1
            k += 1

        while i < len(esquerda):
            lista[k] = esquerda[i]
            i += 1
            k += 1

        while j < len(direita):
            lista[k] = direita[j]
            j += 1
            k += 1

# Teste de execução
dados = [38, 27, 43, 3, 9, 82, 10]
print(f"Lista Original: {dados}")
merge_sort(dados)
print(f"Lista Ordenada: {dados}")
```

####**O que observar:**
<a>Diferente do Quick Sort, o Merge Sort garante um tempo de execução $O(N \log N)$ mesmo no pior cenário possível. Ele consome mais memória (espaço auxiliar), mas oferece estabilidade, mantendo a ordem relativa de elementos iguais.</a>
