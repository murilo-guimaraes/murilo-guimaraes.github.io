##**Introdução Teórica**

A gestão de execução é a forma como o Sistema Operacional organiza o que deve ser processado. Um **Processo** é um programa em execução que possui sua própria "bolha" de memória isolada, garantindo segurança. Já as **Threads** são divisões dentro de um processo que compartilham a mesma memória para realizar tarefas simultâneas de forma mais veloz. O Kernel precisa gerir esses fluxos através de estados (Pronto, Executando ou Bloqueado) e do escalonamento, que decide quem usa a CPU em cada milissegundo. Este relatório demonstra como essa organização permite que o computador realize várias atividades ao mesmo tempo sem travar.

###**1. Simulação de Estados: O Processo Bloqueado**

Um processo entra em estado "Bloqueado" quando precisa esperar por algo externo, como uma resposta do usuário, liberando a CPU para outras tarefas.

```python
import time

print("[ESTADO: EXECUTANDO] Iniciando processamento lógico...")
time.sleep(1)

print("[ESTADO: BLOQUEADO] Aguardando entrada do teclado (I/O)...")
resposta = input("Digite algo para 'DESBLOQUEAR' o processo: ")

print(f"\n[ESTADO: PRONTO] Entrada recebida: '{resposta}'.")
print("[ESTADO: FINALIZADO] Tarefa concluída.")
```

####**O que observar:**
<a>Enquanto o programa espera você digitar, ele não "gasta" processador. O Kernel coloca este processo na fila de espera (Bloqueado) e usa a CPU para outras coisas no Colab. Isso mostra a eficiência do escalonador.</a>

---
###**2. Execução Concorrente com Threads**

Diferente de processos, as threads rodam dentro do mesmo espaço. Vamos rodar duas tarefas ao mesmo tempo para ver quem termina primeiro.

```python
import threading
import time

def tarefa_longa(nome, segundos):
    print(f"Thread {nome}: Iniciada (Duração: {segundos}s)")
    time.sleep(segundos)
    print(f"Thread {nome}: CONCLUÍDA.")

# Criando duas linhas de execução simultâneas
t1 = threading.Thread(target=tarefa_longa, args=("A", 3))
t2 = threading.Thread(target=tarefa_longa, args=("B", 1))

t1.start()
t2.start()

t1.join()
t2.join()
```

####**O que observar:**
<a>Note que a Thread B terminou muito antes da Thread A, mesmo começando quase juntas. O SO gerencia o tempo de cada uma de forma independente, permitindo que tarefas rápidas não fiquem presas atrás de tarefas lentas.</a>

###**3. Identificação de Processos no Sistema**

Cada tarefa no Linux recebe um número único chamado **PID** (Process ID). Vamos verificar o ID do nosso processo atual e de seus "pais".

```python
import os

print(f"ID do Processo Atual (PID): {os.getpid()}")
print(f"ID do Processo Pai (PPID): {os.getppid()}")

# Listando processos ativos que rodam o Python no ambiente
!ps -ef | grep python
```

####**O que observar:**
<a>O PID garante que um processo não mexa nos dados de outro. O comando `ps -ef` mostra que o seu código está rodando dentro de uma estrutura maior gerida pelo Kernel do Google Colab.</a>

###**4. Carga de Trabalho e Escalonamento**

Vamos simular uma carga de processamento para ver como o SO reporta o esforço de execução.

```python
# Verifica a carga de processos nos últimos minutos
print("Carga atual do sistema (Load Average):")
!uptime

# Simula um cálculo rápido para gerar carga
sum(range(10000000))

print("\nCarga após processamento:")
!uptime
```

####**O que observar:**
<a>O `uptime` mostra a média de processos na fila. Se o número aumentar muito, significa que o Kernel está tendo que "fazer malabarismo" para escalonar o tempo de CPU entre todos os pedidos de execução.</a>
