##**Introdução Teórica**

O Sistema Operacional funciona como um gerente entre o utilizador e as peças do computador (hardware). O ponto principal desse sistema é o **Kernel** (núcleo). Para evitar que um erro num programa comum estrague o computador todo, o sistema divide-se em dois modos: o **Modo Usuário**, onde as aplicações comuns rodam com restrições, e o **Modo Kernel**, onde o sistema tem poder total sobre o hardware. Quando um programa precisa de algo do hardware, ele faz um pedido especial chamado **System Call** (Chamada de Sistema). Este relatório explora como essas chamadas funcionam e como o sistema organiza os processadores em arquiteturas modernas.

 - **Monitorar** o gerenciamento de recursos em tempo real para observar o Kernel em ação.
 - **Demonstrar** o isolamento de privilégios através de comandos de sistema que interagem com o núcleo Linux.

###**1. Transição de Modos: Usuário vs Kernel**

Neste primeiro teste, simulamos o tempo que o computador leva para sair do "Modo Usuário" (onde você dá o comando) e entrar no "Modo Kernel" (onde o sistema realmente executa a tarefa no hardware).

```python
import time

def executar_chamada_sistema(comando):
    print(f"[STATUS] Programa em Modo Usuário: Solicitando '{comando}'...")
    time.sleep(0.7) # Simula o processamento inicial
    print(f"[ALERTA] Chaveando para Modo Kernel (Privilegiado)...")
    time.sleep(1.2) # Simula o Kernel acessando o hardware
    print(f"[SUCESSO] Hardware respondeu. Retornando ao Modo Usuário.\n")

executar_chamada_sistema("Gravar ficheiro no disco")
```

####**O que observar:**
<a>Este teste demonstra que nenhum programa toca no hardware diretamente. Existe sempre um "pedágio" (a System Call) que garante que o pedido é seguro antes do Kernel executá-lo. Isso protege o sistema de travamentos totais.</a>

---
###**2. Análise de Processamento Multicore (SMP)**

A maioria dos sistemas modernos usa a arquitetura **SMP** (Symmetric Multi-Processing), onde vários núcleos de CPU dividem a mesma memória. Vamos usar o comando `lscpu` para ver isso na prática.

```python
# Verificando a CPU e a arquitetura do núcleo
!echo "--- Detalhes do Processador ---"
!lscpu | grep -E "CPU\(s\)|Model name|Thread"
```

####**O que observar:**
<a>O resultado mostra quantos núcleos de CPU estão disponíveis. Se houver mais de 1 CPU(s), o Kernel está a gerir uma arquitetura SMP, distribuindo as tarefas de forma equilibrada entre os núcleos para aumentar a velocidade.</a>

---
###**3. Verificação de Memória Disponível**

O Kernel é responsável por decidir quem usa a memória RAM. Vamos verificar como o sistema está a gerir este recurso agora.

```python
# Verificando o estado da memória RAM gerenciada pelo Kernel
!echo "--- Estado da Memória RAM ---"
!free -m
```

####**O que observar:**
<a>A coluna "free" (livre) mostra o que o Kernel ainda tem guardado para novas tarefas. A coluna "buff/cache" mostra a memória que o Kernel está a usar de forma inteligente para acelerar a leitura de ficheiros.</a>

---
###**4. Monitorização de Carga do Sistema**

O Kernel monitoriza constantemente a carga de trabalho. Vamos verificar há quanto tempo o sistema está ligado e qual o nível de esforço médio (Load Average).

```python
# Verificando a carga de trabalho do sistema
!echo "--- Tempo de atividade e Carga ---"
!uptime
```

####**O que observar:**
<a>O "load average" indica se o Kernel está sobrecarregado. Valores baixos significam que o escalonador do SO está a conseguir dar conta de todas as chamadas de sistema sem criar filas de espera.</a>
