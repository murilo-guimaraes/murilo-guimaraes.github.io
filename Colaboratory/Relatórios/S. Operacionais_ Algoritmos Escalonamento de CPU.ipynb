##**Introdução Teórica**

A gerência de processos é uma das funções mais críticas do sistema operacional, sendo o **Escalonador** o componente responsável por decidir qual processo terá acesso à CPU em determinado momento. Segundo Barbosa (2018), o objetivo do escalonamento é maximizar a utilização do processador e garantir que os processos recebam atenção de forma justa. Existem diferentes estratégias para isso: algoritmos não-preemptivos, como o **FIFO** (onde o primeiro a chegar é o primeiro a sair), e algoritmos preemptivos, como o **Round Robin**, que distribui o tempo da CPU em fatias iguais (**quantum**). A escolha do algoritmo impacta diretamente no tempo de espera e no *throughput* do sistema.

###**1. Simulação FIFO: Ordem de Chegada**

O FIFO é o algoritmo mais simples: o processo que solicita a CPU primeiro, a recebe primeiro. É um método não-preemptivo.

```python
import time

def simular_fifo(processos):
    tempo_atual = 0
    print(f"{'Processo':<10} | {'Início':<10} | {'Duração':<10} | {'Fim':<10}")

    for p in processos:
        inicio = tempo_atual
        time.sleep(p['duracao'] / 10) # Simulação acelerada
        tempo_atual += p['duracao']
        print(f"{p['nome']:<10} | {inicio:<10} | {p['duracao']:<10} | {tempo_atual:<10}")

lista_processos = [
    {'nome': 'P1', 'duracao': 10},
    {'nome': 'P2', 'duracao': 2},
    {'nome': 'P3', 'duracao': 4}
]

simular_fifo(lista_processos)
```

####**O que observar:**
<a>No FIFO, se um processo longo (P1) chegar primeiro, ele bloqueia a CPU para os processos curtos (P2 e P3), aumentando o tempo médio de espera. Este fenômeno é conhecido tecnicamente como **Efeito Comboio**.</a>

---
###**2. Round Robin: Compartilhamento por Tempo**

No Round Robin, cada processo recebe uma fatia de tempo (Quantum). Se não terminar nesse tempo, ele volta para o fim da fila.

```python
def simular_round_robin(processos, quantum):
    fila = list(processos)
    tempo = 0
    while fila:
        p = fila.pop(0)
        execucao = min(p['restante'], quantum)
        p['restante'] -= execucao
        tempo += execucao
        print(f"Tempo {tempo}: {p['nome']} executou por {execucao}u. (Restante: {p['restante']})")

        if p['restante'] > 0:
            fila.append(p)

processos_rr = [
    {'nome': 'P1', 'restante': 10},
    {'nome': 'P2', 'restante': 4},
    {'nome': 'P3', 'restante': 2}
]

simular_round_robin(processos_rr, quantum=3)
```

####**O que observar:**
<a>Diferente do FIFO, o Round Robin impede que um único processo monopolize a CPU. Através da preempção, o sistema simula a execução simultânea, essencial para sistemas interativos. O tamanho do **Quantum** é crítico: se for muito pequeno, o SO perde muito tempo trocando de contexto.</a>

###**3. Diagnóstico de Sistema: Valor 'Nice'**

Em sistemas reais (como o kernel Linux do Colab), processos possuem prioridades definidas pelo valor **Nice**. Vamos verificar os processos ativos no container.

```python
# Verificando processos e suas prioridades (NI = Nice Value)
!ps -el | head -n 10
```

####**O que observar:**
<a>A coluna **NI** mostra o valor Nice. Valores menores indicam maior prioridade para o escalonador. Como ensina Barbosa, o escalonamento por prioridade permite que o SO dê preferência a tarefas críticas do sistema ou interações do usuário.</a>

##**Conclusão**

A simulação dos algoritmos de escalonamento permitiu compreender como o sistema operacional equilibra a carga de trabalho. Enquanto o **FIFO** prioriza a simplicidade, ele falha em eficiência quando há grande variação no tempo de execução dos processos. O **Round Robin** demonstrou ser a solução ideal para a multiprogramação moderna, garantindo fluidez ao sistema através da preempção.
