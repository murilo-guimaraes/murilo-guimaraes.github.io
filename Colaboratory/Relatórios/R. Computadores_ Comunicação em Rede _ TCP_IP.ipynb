##**Introdução Teórica**

O modelo **TCP/IP** é a base de comunicação da internet moderna. Ele organiza o envio de informações em quatro camadas principais: **Aplicação**, **Transporte**, **Internet** e **Acesso à Rede**. Diferente de outros modelos, o TCP/IP foca na eficiência e na garantia de que os dados cheguem ao destino correto através de endereços IP e portas de comunicação. Como o ambiente de nuvem bloqueia protocolos de teste simples (ICMP/Ping), utilizamos ferramentas de diagnóstico de porta para validar a conexão na camada de Transporte. Este relatório explora como essas camadas interagem para garantir que um pedido feito pelo usuário chegue com segurança ao servidor.

---
##**Objetivos**

 - **Identificar** as interfaces de rede e o endereço IP atribuído à máquina virtual.
 - **Verificar** a disponibilidade de serviços remotos através do escaneamento de portas (TCP).
 - **Rastrear** o caminho dos pacotes (saltos) até o destino final na camada de Internet.
 - **Validar** a entrega de dados na camada de Aplicação através de uma requisição HTTP.

---
###**1. Verificação de Interface e Endereço IP**

O primeiro passo é saber quem somos na rede. Vamos listar as interfaces e os IPs da máquina.

```python
# Listando as configurações de rede do container Linux
!ip addr show eth0
```

####**O que observar:**
<a>O comando exibe o IP interno da máquina. Isso confirma que a **Camada de Internet** está ativa e que o dispositivo possui um endereço lógico para enviar e receber pacotes dentro da rede do Google.</a>

---
###**2. Teste de Serviço Remoto (Substituto do Ping)**

Como o `ping` é bloqueado, usamos o `nmap` para ver se a "porta" de um servidor (como a 443 do Google) está aberta e aceitando conexões.

```python
# Instalando uma ferramenta leve de diagnóstico
!apt-get install nmap -y > /dev/null

# Verificando se a porta HTTPS (443) do Google está acessível
!nmap -p 443 google.com
```

####**O que observar:**
<a>Se o estado aparecer como `open`, a **Camada de Transporte** (TCP) conseguiu estabelecer uma conexão. Isso prova que o servidor está "vivo" e acessível, contornando o bloqueio do protocolo ICMP.</a>

---
###**3. Rastreio de Saltos (Hop-by-Hop)**

Mesmo com as restrições de segurança do ambiente Cloud, podemos usar o `traceroute` para mapear por quantos roteadores a informação passa até chegar ao destino final.

```python
# Instalando o traceroute e executando o rastreio de rota
!apt-get install traceroute -y > /dev/null
!traceroute -n 8.8.8.8
```

####**O que observar:**
<a>Cada linha numerada representa um roteador (salto) no caminho. </a>

><a>**Nota técnica:** No Google Colab, é comum aparecerem asteriscos (`* * *`) em alguns saltos; isso ocorre porque os firewalls da infraestrutura do Google bloqueiam a resposta de IPs internos por segurança. O teste é bem-sucedido ao mostrar a transição da rede local para a internet (Camada de Internet).</a>

---
###**4. Comunicação e Resposta HTTP**

Neste teste final, simulamos o acesso a um site para ver a resposta que o servidor envia para a nossa aplicação.

```python
# Solicitando apenas o cabeçalho de uma página web
!curl -I https://www.wikipedia.org
```

####**O que observar:**
<a>O log confirma o sucesso da comunicação na **Camada de Aplicação**. O código `HTTP/2 200` indica que o servidor aceitou o pedido. Podemos identificar o tipo de conteúdo entregue (`text/html`), o servidor utilizado (`ATS/9.2.13`) e até o IP de origem registrado (`34.24.205.159`). Isso prova que, após passar pelas camadas de Internet e Transporte, os dados foram entregues e interpretados corretamente pelo protocolo de aplicação.</a>
