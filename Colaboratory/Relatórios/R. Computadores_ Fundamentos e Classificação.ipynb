##**Introdução Teórica**

De acordo com Nunes (2017), uma rede de computadores é um conjunto de módulos processadores capazes de trocar informações e compartilhar recursos, interconectados por um sistema de comunicação. A classificação dessas redes ocorre principalmente pela sua abrangência geográfica, variando de **LAN** (Local), **MAN** (Metropolitana) até **WAN** (Longa Distância). Além da escala, a eficiência da rede é determinada pela sua **Topologia** (layout físico e lógico) e pelo **Modo de Transmissão** (Simplex, Half-Duplex ou Full-Duplex). Compreender esses fundamentos é o primeiro passo para o projeto de arquiteturas que suportem o tráfego de dados global de forma resiliente.

###**1. Diagnóstico de Alcance de Rede**

Vamos identificar o endereço IP local e compará-lo com o IP público da rede. Isso nos ajuda a entender a transição da rede local (LAN) para a rede de longa distância (WAN).

```python
# Verificando o IP Interno (LAN do Data Center)
print("--- Endereço IP Interno (LAN) ---")
!hostname -I

# Verificando o IP Externo (Saída para a WAN)
print("\n--- Endereço IP Público (WAN) ---")
!curl -s https://ifconfig.me
```

####**O que observar:**
<a>A existência de IPs distintos confirma que a máquina está em uma **LAN** privada que se comunica com a **WAN** (Internet) através de um gateway. Como ensina Nunes, a WAN interconecta múltiplas LANs através de distâncias globais.</a>

---
###**2. Latência e Abrangência Geográfica**

Embora o ICMP seja restrito, podemos medir o tempo de conexão TCP com servidores em diferentes partes do mundo para entender o impacto da distância em uma WAN.

```python
import time
import socket

def testar_latencia_tcp(host, porta=80):
    start = time.time()
    try:
        sock = socket.create_connection((host, porta), timeout=5)
        end = time.time()
        sock.close()
        print(f"Tempo de resposta para {host}: {(end - start)*1000:.2f} ms")
    except:
        print(f"Falha ao conectar em {host}")

print("Testando resposta (Custo de WAN):")
testar_latencia_tcp("google.com.br")
testar_latencia_tcp("www.u-tokyo.ac.jp") # Servidor no Japão
```

####**O que observar:**
<a>A diferença nos tempos de resposta ilustra a escala da **WAN**. Servidores fisicamente mais distantes apresentam maior latência devido ao número de saltos e à velocidade de propagação nos meios de transmissão intercontinentais.</a>

###**3. Modo de Operação da Interface**

Nesta etapa, verificamos as propriedades da interface `eth0` para entender como o sistema operacional lida com o fluxo de dados (Full-Duplex por padrão em sistemas modernos).

```python
# Verificando estatísticas de transmissão e recepção (TX/RX)
!ip -s link show eth0
```

####**O que observar:**
<a>Os contadores de pacotes transmitidos (TX) e recebidos (RX) operando simultaneamente indicam uma operação **Full-Duplex**. Conforme o livro de Nunes, isso permite que o host envie e receba dados ao mesmo tempo, dobrando a eficiência do canal físico.</a>
